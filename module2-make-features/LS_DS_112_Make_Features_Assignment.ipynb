```python
# I want to grab and mess the intrest rates but as they are I can't do anything since they are objects. I'm going to check a specific column to see exactly how it was typed 

df['int_rate'][5]
```

```python
# Noticed that there was a space in front of the object, now I'm going to check the very first index to make sure it's consistant with single digit numbers 

df['int_rate'][0]
```

```python
# Now this is intresting, for single digit numbers it's two spaces and double digit numbers it is one space, going to make a function to elimiate the spaces and % and convert the values into floats 

def Object_to_float(int_rate):
  return float(int_rate.strip().strip('%'))
```

```python
# I'm going to do a test to make sure my function works just fine 

test = ' 54.421%'

test1 = Object_to_float(test)

print(test1)
print(type(test1))
```

```python
# My function works great, now using the .apply funtion I am going to add a new column in the end with the updated numbers, leaving the original column so I won't forget they were percentages

df['int_rate_float'] = df['int_rate'].apply(Object_to_float)
```

```python
# Now I'm going to look at the employment to make sure all the names match in capitalization and spacing
# Showing the first 20 most common job titles, also showing the NaN values so I can figure out how many are not filled out 

df['emp_title'].value_counts(dropna=False).head(20)
```

```python
# Knowing that these are the most common job titles, I'm going to check how many unique jobs are listed. 
df['emp_title'].nunique()
```

```python
# I now know that there's way too many unique values and there has to be similar job titles but space or capitilized wrong
# I'll write a function to make sure everything is uniform. 

def clean_job_title(title):
  if isinstance(title, str):
    return title.strip().title()
  else:
    return "Unknown"
```

```python
# I'm going to test the function to make sure it properlly works 

ex = ['job', ' jobless', ' Looking for job', np.NaN]

for example in ex:
  print(clean_job_title(example))

'''
Leaving this as my original, had a bit of trouble where everything was returning Uknown, funny thing 
was that it was the fact it was returning all Unknowns

for example in ex:
  print(clean_job_title(ex))
'''
```

```python
# Now knowing that the funciton works, I could apply it to the column and it would unify all the data to the same format

df['emp_title'] = df['emp_title'].apply(clean_job_title)
```

```python
# Ran the code, now checking to see if values have updated

df['emp_title'].value_counts().head(20)
```

```python
# Already seeing a huge jump in matching job titles, now to see how many unique job titles are there now 

df['emp_title'].nunique()
```
